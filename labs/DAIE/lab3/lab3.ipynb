```python
import pandas as pd
import numpy as np
```

```python
data = {
    'emp_id': range(1, 12),
    'first_name': ['John', 'Jane', 'Bob', 'Alice', 'Eve', 'Charlie', 'David', 'Emily', 'Frank', 'Grace', "Sahil"],
    'last_name': ['Doe', 'Smith', 'Johnson', 'Brown', 'Jones', 'Wilson', 'Miller', 'Davis', 'Taylor', 'Clark', "Sapariya"],
    'gender': ['M', 'F', 'M', 'F', 'F', 'M', 'M', 'F', 'M', 'F', 'M'],
    'age': np.random.randint(22, 60, 11),
    'salary': np.random.randint(50000, 100000, 11),
    'department_id': np.random.randint(1, 5, 11)
}
```

```python
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df['gender'] = df['gender'].replace({ "M": "Male", "F": "Female" })
df
```

```python
df.head(3) # without any argument in head() and tail() will give default 5 rows
```

```python
df.tail(4)
```

```python
df["last_name"] = df['last_name'].str.replace('Doe', 'Cena')
df
```

```python
females = df[df['gender'].str.contains('Female')]
females
```

```python
males = df[df['gender'].str.contains('Male')]
males
```

```python
age_30 = df[df['age'] > 30]
age_30
```

```python
temp = df
temp['age'] = np.where(df['age'] > 30, np.nan, df['age'])
temp
```

```python
df['bonus'] = 0.1 * df['salary']
df
```

```python
gender_group = df.groupby('gender')
gender_group
```

```python
gender_group.get_group('Male')
```

```python
gender_male_salary_mean = gender_group.get_group('Male')['salary'].mean()
gender_male_salary_mean
```

```python
group_salary_mean = gender_group['salary'].mean()
group_salary_mean
```

```python
df['department_name'] = np.select(
    [
        (df['department_id'] == 1),
        (df['department_id'] == 2),
        (df['department_id'] == 3),
        (df['department_id'] == 4),
        (df['department_id'] == 5),
    ],
    [
        "IT",
        "Finance",
        "Marketing",
        "Mining",
        "Networking"
    ],
    default=None # Default value if none of the conditions is met
)
df
```

```python
df["full_name"] = df["first_name"] + " " + df['last_name']
df
```

```python
pivot_df = df.pivot_table(index='department_name', values='salary', aggfunc=np.mean)
pivot_df
```
