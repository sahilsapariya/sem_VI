__Aim:__ To perform Navie Bayes and K-NN (k-nearest neighbor) classification using sklearn package in python

```python
# Naive Bayes
# -------------
# P(A/B) = P(B/A) x P(A) / P(B)
# Hypothesis means assumptions
# A -> Hypothesis
# B -> Data
# P(A/B) -> Posterier Probability (Hypothesis given the data)
# P(B/A) -> Posterier Probability (Data given the Hypothesis)
# P(A) -> Prior Probbability (Of Hypothesis)
# P(B) -> Prior Probability ( Of Given Data)

# Example
# ----------
# Weather is our target
# Play or not is our class
#  A -> Play
#  B -> Weather , Temprature, Wind, Humidaity (It will be in fixed no of partitions)

# Example Table
# Weather	Play
# --------------
# Sunny	    No  
# Sunny	    No
# Overcast	Yes
# Rainy	    Yes
# Rainy	    Yes
# Rainy	    No
# Overcast	Yes
# Sunny	    No
# Sunny	    Yes
# Rainy	    Yes
# Sunny	    Yes
# Overcast	Yes
# Overcast	Yes
# Rainy	    No
```

```python
# import Statements
from sklearn.model_selection import train_test_split as tts
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris
from sklearn.metrics import accuracy_score,confusion_matrix
import matplotlib.pyplot as pl
%matplotlib inline
```

```python
# Data
iris = load_iris()
```

```python
X= iris.data
Y=iris.target
```

```python
# Train test split
x_train,x_test,y_train,y_test = tts(X,Y,train_size=0.7,random_state=42)
```

```python
# Classifier variable
gnb=GaussianNB()
```

```python
# Fit the model
gnb.fit(x_train,y_train)
```

```python
## We Need to predicition
y_prediction = gnb.predict(x_test)
```

```python
## 
print(accuracy_score(y_prediction,y_test))
```

```python
cm = confusion_matrix(y_test,y_prediction)
cm
```

```python
pl.plot(cm)
```
