__Aim:__ To perform KMeans and DBSCAN clustering using sklearn package in python

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN
```

```python
X, y_true = make_blobs(n_samples=100, centers=3, cluster_std=0.6)
```

```python
km = KMeans(n_clusters=3)
```

```python
km.fit(X)
y_lab = km.predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_lab, s=50, cmap='viridis')
centers = km.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.title('KMeans Clustering')
plt.show()
```

```python
dbs = DBSCAN()
```

```python
y_dbs = dbs.fit_predict(X)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y_dbs, s=50, cmap='viridis')
plt.title('DBSCAN Clustering')
plt.show()
```
