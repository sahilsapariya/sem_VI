# Experiment 4

To perform Data Visualization and plotting techniques like Line Plot, bar-chart, pie-chart, box-plot using Matplotlib libraries

```python
import matplotlib.pyplot as plt
import numpy as np
import sklearn.datasets as data
import pandas as pd
import seaborn as sns
```

```python
iris = data.load_iris()
scatter = plt.scatter(x=iris.data[:, 0], y=iris.data[:,1])
```

```python
plt.hist(iris.data[:, 0], 20)
```

```python
df = pd.DataFrame(data=np.c_[iris['data'], iris['target']], columns=iris['feature_names'] + ['target'])
df
```

```python
df['target'] = df['target'].astype(int)
df
```

```python
# Set the style of seaborn
sns.set(style="whitegrid")
```

```python
# Plot bar charts for each feature based on target classes

# Bar chart for Sepal Length
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
sns.barplot(x='target', y='sepal length (cm)', data=df, errorbar=None)
plt.title('Bar chart for Sepal Length')

# Bar chart for Sepal Width
plt.subplot(2, 2, 2)
sns.barplot(x='target', y='sepal width (cm)', data=df, errorbar=None)
plt.title('Bar chart for Sepal Width')

# Bar chart for Petal Length
plt.subplot(2, 2, 3)
sns.barplot(x='target', y='petal length (cm)', data=df, errorbar=None)
plt.title('Bar chart for Petal Length')

# Bar chart for Petal Width
plt.subplot(2, 2, 4)
sns.barplot(x='target', y='petal width (cm)', data=df, errorbar=None)
plt.title('Bar chart for Petal Width')

plt.tight_layout()
plt.show()
```

# Barchart for Iris datasets without using seaborn

```python
# Bar chart for Sepal Length
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
bar_width = 0.8
bar_colors = ['blue', 'orange', 'green']
for i in range(3):
    subset = df[df['target'] == i]
    plt.bar(subset['target'] + i * bar_width / 3, subset['sepal length (cm)'], width=bar_width/3, color=bar_colors[i])
plt.title('Bar chart for Sepal Length')
plt.xticks([0, 1, 2], labels=['Setosa', 'Versicolor', 'Virginica'])

# Bar chart for Sepal Width
plt.subplot(2, 2, 2)
for i in range(3):
    subset = df[df['target'] == i]
    plt.bar(subset['target'] + i * bar_width / 3, subset['sepal width (cm)'], width=bar_width/3, color=bar_colors[i])
plt.title('Bar chart for Sepal Width')
plt.xticks([0, 1, 2], labels=['Setosa', 'Versicolor', 'Virginica'])

# Bar chart for Petal Length
plt.subplot(2, 2, 3)
for i in range(3):
    subset = df[df['target'] == i]
    plt.bar(subset['target'] + i * bar_width / 3, subset['petal length (cm)'], width=bar_width/3, color=bar_colors[i])
plt.title('Bar chart for Petal Length')
plt.xticks([0, 1, 2], labels=['Setosa', 'Versicolor', 'Virginica'])

# Bar chart for Petal Width
plt.subplot(2, 2, 4)
for i in range(3):
    subset = df[df['target'] == i]
    plt.bar(subset['target'] + i * bar_width / 3, subset['petal width (cm)'], width=bar_width/3, color=bar_colors[i])
plt.title('Bar chart for Petal Width')
plt.xticks([0, 1, 2], labels=['Setosa', 'Versicolor', 'Virginica'])

plt.tight_layout()
plt.show()
```
